# Deep EDA with DissectML

Explore all `iml.explore()` sub-modules.

```python
import dissectml as iml

df = iml.load_titanic()
eda = iml.explore(df)
# explore() returns instantly — all computation is lazy
```

## Overview

```python
eda.overview.show()
```

## Correlations

```python
# Unified matrix: auto-selects statistically appropriate measure per cell
eda.correlations.heatmap().show()
```

## Missing Data Analysis

```python
eda.missing.summary()
```

## Outlier Detection

```python
# Consensus: flagged by ≥2 of IQR / Z-score / Isolation Forest
eda.outliers.consensus(min_methods=2)
```

## Statistical Tests

```python
eda.tests.normality()
```

## Auto Clustering

```python
eda.clusters.plot().show()
```
