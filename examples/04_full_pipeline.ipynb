# Full Pipeline — End to End

This notebook walks through the complete DissectML pipeline:
EDA → Intelligence → Battle → Compare → Report

Using the California Housing regression dataset.

```python
import dissectml as iml

df = iml.load_housing()
print(f"Shape: {df.shape}")
print(f"Target: MedHouseVal")
df.head()
```

## Stage 1 — EDA

```python
eda = iml.explore(df)
eda.overview.show()
```

```python
eda.correlations.heatmap().show()
```

## Stage 2 — Intelligence Bridge

```python
intel = iml.analyze_intelligence(df, target="MedHouseVal", task="regression")

print(f"Readiness score: {intel.readiness.score}/100 (grade {intel.readiness.grade})")
print(f"Leakage warnings: {len(intel.leakage)}")
```

```python
intel.readiness.waterfall_figure().show()
```

```python
intel.feature_importance.sort_values("composite_rank").head(8)
```

## Stage 3 — Model Battle

```python
models = iml.battle(
    df,
    target="MedHouseVal",
    task="regression",
    families=["linear", "tree", "ensemble"],
    cv=5,
    n_jobs=-1,
)
models.leaderboard()
```

## Stage 4 — Compare

```python
comp = iml.ModelComparator(models, y=df["MedHouseVal"])
comp.residual_plots.show()
```

```python
comp.pareto.show()
```

## Stage 5 — Full Report

```python
report = iml.analyze(
    df,
    target="MedHouseVal",
    task="regression",
    battle_families=["linear", "tree", "ensemble"],
    cv=5,
    n_jobs=-1,
)

print(report.summary())
```

```python
out = report.export("housing_report.html")
print(f"Full report saved to: {out}")
```
