# DissectML Quickstart

This notebook demonstrates the full `iml.analyze()` pipeline in under 5 minutes.

```python
import dissectml as iml

print(f"DissectML version: {iml.__version__}")
```

## 1. Load a demo dataset

```python
df = iml.load_titanic()
print(f"Shape: {df.shape}")
df.head()
```

## 2. Run the full pipeline

```python
report = iml.analyze(
    df,
    target="survived",
    task="classification",
    battle_families=["linear", "tree"],
    cv=5,
    n_jobs=-1,
)
```

## 3. Inspect results

```python
print(report.summary())
```

```python
report.models.leaderboard()
```

```python
report.compare.pareto.show()
```

## 4. Export the report

```python
out = report.export("titanic_report.html")
print(f"Report saved to: {out}")
```
